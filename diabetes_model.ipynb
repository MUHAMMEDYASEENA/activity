```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data=pd.read_csv(r'C:\Users\welcome\Downloads\diabetes_model.csv')
data
```

# Count plot of Outcome

```python
sns.countplot(data['Outcome'])
```

# Analyse effect of age on Outcome using Boxplot

```python
sns.boxplot(x="Outcome",y="Age",data=data)
```

# Check for the chances of getting diabetes if it is common in family, using the Diabetes Pedigree Function

```python
sns.swarmplot(y=data['DiabetesPedigreeFunction'],x=data['Outcome'],data=data)
```

# Check for the relationship with Insulin & Glucose and group the datapoints by Outcome

```python
plt.figure(figsize=(8,8))
sns.stripplot(x=data['Insulin'],y=data['Glucose'],data=data,hue=data['Outcome'])
```

# Create a heatmap on Correlation of features in the data

```python
sns.heatmap(data.corr(),annot=True)
```
